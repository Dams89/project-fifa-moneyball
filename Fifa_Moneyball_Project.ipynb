```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fifa = pd.read_csv('fifa21_male2.csv')
fifa.head()
```

```python
fifa['Height'] = fifa['Height'].str.replace('"','')
```

```python
conversions = [30.48, 2.54]
fifa['Height'] = fifa['Height'].str.split("'").apply(pd.Series).astype(int).dot(conversions)
fifa.head()
```

```python
fifa = fifa[fifa['Position']=='GK']
fifa.head()
```

```python
fifa = fifa[['ID','Name','Age','Height','OVA', 'POT' ,'Goalkeeping','GK Diving','GK Handling','GK Kicking','GK Positioning','GK Reflexes','Total Stats','Base Stats','IR']]
fifa.head() 
```

```python
fifa['IR'] = fifa['IR'].str.replace('★','').astype('int')
fifa.head()
```

```python
sns.heatmap(fifa.corr(),annot=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

normalizer = MinMaxScaler()
```

```python
fifa['OVA'] = normalizer.fit_transform(fifa[['OVA']])
fifa['Goalkeeping'] = normalizer.fit_transform(fifa[['Goalkeeping']])
fifa['IR'] = normalizer.fit_transform(fifa[['IR']])
fifa['Total Stats'] = normalizer.fit_transform(fifa[['Total Stats']])
fifa['Base Stats'] = normalizer.fit_transform(fifa[['Base Stats']])
fifa['GK Diving'] = normalizer.fit_transform(fifa[['GK Diving']])
fifa['GK Handling'] = normalizer.fit_transform(fifa[['GK Handling']])
fifa['GK Kicking'] = normalizer.fit_transform(fifa[['GK Kicking']])
fifa['GK Positioning'] = normalizer.fit_transform(fifa[['GK Positioning']])
fifa['GK Reflexes'] = normalizer.fit_transform(fifa[['GK Reflexes']])
fifa['Height'] = normalizer.fit_transform(fifa[['Height']])
fifa = pd.get_dummies(fifa, columns=['Age'], drop_first=True)
```

```python
fifa = fifa[['ID','Age_17','Age_18','Age_19','Age_20','Age_21','Age_22','Age_23','Age_24','Age_25','Age_26','Age_27','Age_28','Age_29','Age_30','Age_31','Age_32','Age_33','Age_34','Age_35','Age_36','Age_37','Age_38','Age_39','Age_40','Age_41','Age_42','Age_43','Age_47','Height','OVA', 'POT' ,'Goalkeeping','GK Diving','GK Handling','GK Kicking','GK Positioning','GK Reflexes','Total Stats','Base Stats','IR']]
```

```python
X=fifa.drop(columns=['POT'])
y=fifa['POT']
```

```python
from sklearn import linear_model

model = linear_model.LinearRegression()
result = model.fit(X,y)
```

```python
result.score(X,y)
```
